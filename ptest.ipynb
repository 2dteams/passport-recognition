```python
import numpy as np
import cv2 as cv

#   
hsv_min = np.array((4, 30, 17), np.uint8)
hsv_max = np.array((120, 85, 60), np.uint8)

img = cv.imread("333.jpg")
img = cv2.resize(img, (0,0), fx=0.8, fy=0.8)
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV )
#     BGR  HSV
thresh = cv.inRange(hsv, hsv_min, hsv_max )
#   
#        contours
contours, hierarchy = cv.findContours(thresh.copy(), cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
# hierarchy    
#    
cv.drawContours(img, contours, -1, (255, 0, 0), 2, cv.LINE_AA, hierarchy, 0)
cv.imshow('contours', img)
cv.drawContours(img, contours, -1, (255, 0, 0), 2, cv.LINE_AA, hierarchy, 1)
cv.drawContours(img, contours, -1, (255, 0, 0), 2, cv.LINE_AA, hierarchy, 2)

text = pytesseract.image_to_string(img, lang='rus').split('\n')
for i in text:
    print(i)
#     
cv.imshow('All_con', img)
# pyplot.imshow(thresh)
# cv.imshow('thresh', thresh)
cv.waitKey()
cv.destroyAllWindows()
```
